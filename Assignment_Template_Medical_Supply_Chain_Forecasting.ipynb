```python
# import pandas and statsmodels
import pandas as pd
import statsmodels.api as sm
```

```python
# import the data

bitCases = pd.read_csv('Assignment_Data_Medical_Supply_Chain_Forecasting.csv')

# Use .head(10) to check the data
bitCases.head(10)
```

```python
# Set X as Date and Temperature and y as Cases
X = bitCases[['Date','Temperature']]
y = bitCases['Cases']
```

```python
# Fit the regression and print the results

# Add the constant to X
X = sm.add_constant(X)

# Fit the data
model = sm.OLS(y,X).fit()

# Print the model summary
print_model = model.summary()
print(print_model)
```
