```python
# Import pandas, sklearn, train_test_split, KNeighborsClassifier, and accuracy_score
import pandas as pd
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import KNNImputer
```

```python
# Part 1
```

```python
# Import the data with "Facility Type" values
df = pd.read_csv('Is_This_A_Hospital_Or_Nursing_Home_Training_Test_Set_2.csv')

# Use .head(10) to check your data
df.head(10)
```

```python
# Set the X and y. Be sure to use the normalized values for X.
le = LabelEncoder()
label = le.fit_transform(df['Facility Type'])

df.drop("Facility Type", axis=1, inplace=True)
df["Facility Type"] = label
df.head(5)
X = df.drop(['Name', 'Average Patient Age', 'Number of Patients Served per Year', 'Average Patient Cost','Average Age of Equipment'], axis=1)
y=df['Facility Type']
# Use .head() to ensure the data was replaced with numerals
```

```python
# Use train_test_split to randomly partition your data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)


# Describe Xtrain to ensure there is not a skew on your data (mean should be close to 0 since it is normalized)
```

```python
# Part 2
```

```python
# Set the KNN setting to n_neighbors=1
knn = KNeighborsClassifier(n_neighbors = 1)

# Fit the data
knn.fit(X_train,y_train)
```

```python
# Set the predictions to a series
knn.predict(X_test)

# Compare the predictions to the actual values
knn.score(X_test, y_test)
```

```python
# Use k=3

knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train,y_train)
pred3 = knn3.predict(X_test)
accuracy_score(pred3,y_test)
```

```python
# Use k=5
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train,y_train)
pred5 = knn5.predict(X_test)
accuracy_score(pred5,y_test)
```

```python
# Use k=7
knn7 = KNeighborsClassifier(n_neighbors=7)
knn7.fit(X_train,y_train)
pred7 = knn7.predict(X_test)
accuracy_score(pred7,y_test)
```

```python
# Use k=9
knn9 = KNeighborsClassifier(n_neighbors=9)
knn9.fit(X_train,y_train)
pred9 = knn9.predict(X_test)
accuracy_score(pred9,y_test)
```

```python
# Use k=11
knn11 = KNeighborsClassifier(n_neighbors=11)
knn11.fit(X_train,y_train)
pred11 = knn11.predict(X_test)
accuracy_score(pred11,y_test)
```
